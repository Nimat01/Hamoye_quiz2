```python
df_new=df.drop(['date', 'lights'], axis=1)
```

```python
scaler=MinMaxScaler()
norm_df=scaler.fit_transform(df_new)
norm_df=pd.DataFrame(norm_df, columns=df_new.columns)
norm_df.head()
```

```python
x_new=norm_df.drop('Appliances', axis=1)
y_new=norm_df['Appliances']
```

```python
X_train, X_test, y_train, y_test=train_test_split(x_new, y_new, test_size=0.3, random_state=42)
```

```python
linreg=LinearRegression()
linreg.fit(X_train, y_train)
y_pred=linreg.predict(X_test)
MAE=mean_absolute_error(y_test, y_pred)
RMSS=np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
ridge=Ridge(0.4)
ridge.fit(X_train, y_train)
ridge_pred=ridge.predict(X_test)
```

```python
mean_absolute_error(y_test, ridge_pred)
```

```python
lasso_reg=Lasso(alpha=0.001)
lasso_reg.fit(X_train, y_train)
lasso_pred=lasso_reg.predict(X_test)
np.sqrt(mean_squared_error(y_test, lasso_pred))
```

```python
def get_weights(model, fit, col_name):
    weights=pd.Series(model.coef_, fit.columns).sort_values()
    weights_df=pd.DataFrame(weights).reset_index()
    weights_df.columns=['Feautures', col_name]
    weights_df[col_name].round(2)
    return weights_df
```

```python
model_weight=get_weights(linreg, X_train, 'model_weights')
ridge_weight=get_weights(ridge, X_train, 'ridge_weights')
lasso_weight=get_weights(lasso_reg, X_train, 'lasso_weights')
```

```python
final1=pd.merge(model_weight, ridge_weight, on='Feautures')
final2=pd.merge(final1, lasso_weight, on='Feautures')
```
